# import dataset and convert into csv/dataframe

```python
def parse(path):
  g = gzip.open(path, 'rb')
  for l in g:
    yield eval(l)

def getDF(path):
  i = 0
  df = {}
  for d in parse(path):
    df[i] = d
    i += 1
  return pd.DataFrame.from_dict(df, orient='index')

df_meta = getDF('meta_Sports_and_Outdoors.json.gz')
df_reviews = getDF('reviews_Sports_and_Outdoors_5.json.gz')
```

## Populate nodes and edges

```python
#node csv
nodes = []
[nodes.append(node) for node in df_meta_copy['asin']]
[nodes.extend(node) for node in df_meta_copy['also_bought'] if node != None]
[nodes.extend(node) for node in df_meta_copy['also_viewed'] if node != None]
[nodes.extend(node) for node in df_meta_copy['bought_together'] if node != None]

#nodes
df_nodes = pd.DataFrame(nodes, columns = ['Nodes'])
df_nodes.shape
```

```python
df_nodes.head()
```

```python
#edges
temp_list_1 = []
for i in range(len(df_meta_copy['asin'])):
    if df_meta_copy['also_bought'].iloc[i] != None:
        for e in range(len(df_meta_copy['also_bought'].iloc[i])):
            temp_list_2 = []
            temp_list_2.append(df_meta_copy['asin'].iloc[i])
            temp_list_2.append(df_meta_copy['also_bought'].iloc[i][e])
            temp_list_2.append(2)
            temp_list_1.append(temp_list_2)
    if df_meta_copy['also_viewed'].iloc[i] != None:
        for e in range(len(df_meta_copy['also_viewed'].iloc[i])):
            temp_list_2 = []
            temp_list_2.append(df_meta_copy['asin'].iloc[i])
            temp_list_2.append(df_meta_copy['also_viewed'].iloc[i][e])
            temp_list_2.append(3)
            temp_list_1.append(temp_list_2)
    if df_meta_copy['bought_together'].iloc[i] != None:
        for e in range(len(df_meta_copy['bought_together'].iloc[i])):
            temp_list_2 = []
            temp_list_2.append(df_meta_copy['asin'].iloc[i])
            temp_list_2.append(df_meta_copy['bought_together'].iloc[i][e])
            temp_list_2.append(1)
            temp_list_1.append(temp_list_2)
    
```

## Update weights with reviews

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
df_reviews = pd.read_json('reviews.json')
```

```python
df_reviews.head(20)
df_reviews = df_reviews.drop(['reviewerID', 'reviewerName', 'reviewText', 'summary', 'unixReviewTime', 'reviewTime'], axis = 1)
```

```python
df_reviews.head()
```

```python
helpful_fraction = []

for i in range(len(df_reviews.helpful)):
    if df_reviews.helpful.iloc[i][1] != 0:
        calc = df_reviews.helpful.iloc[i][0]/df_reviews.helpful.iloc[i][1]
        helpful_fraction.append(calc)
    else: helpful_fraction.append(0)
```

```python
df_reviews['helpful_calc'] = helpful_fraction
df_reviews = df_reviews.drop(['helpful'], axis = 1)
```

```python
#scaler fit transform
df_reviews[['overall', 'helpful_calc']] = scaler.fit_transform(df_reviews[['overall', 'helpful_calc']])
df_reviews.head()
```

```python
alpha = 0.5
overall_score = [row for row in df_reviews.overall]
helpful_calc = [row for row in df_reviews.helpful_calc]

additive_score = []
for i in range(len(overall_score)):
    score = (overall_score[i] * alpha) + ((1-alpha) * helpful_calc[i])
    additive_score.append(score)
```

```python
df_reviews['additive_score'] = additive_score
```

```python
df_reviews = df_reviews.drop(['overall', 'helpful_calc'], axis=1)
df_reviews_group = df_reviews.groupby('asin').mean()
df_reviews_group.reset_index(inplace=True)
```

```python
df_final['new_weight'] = df_final['weight'] * df_final['additive_score_x'] * df_final['additive_score_y']
```

```python
df_edges_final = df_final_1.groupby(['Source', 'Target']).mean()
```
